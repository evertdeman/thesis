```python
import pandas as pd 
import numpy as np
import glob
from os.path import join 
import os 
from scikits.bootstrap import ci

from vis import Visualization, parse_table, pbcc_results
```

```python
def make_dfs(dir_id, dir_pr, pt=True):
    
    inmod = dir_id.split("/")[2].split("_")[0]
    
    df_run = pd.DataFrame()
    df_run = df_run.append(pd.read_csv(join("..", dir_pr, "run.csv")).assign(setting="pr").assign(inmod=inmod))
    df_run = df_run.append(pd.read_csv(join("..", dir_id, "run.csv")).assign(setting="id").assign(inmod=inmod))

    if pt:
        df_pt = pd.DataFrame()
        df_pt = df_pt.append(pd.read_csv(join("..", dir_pr, "pt.csv")).assign(setting="pr").assign(inmod=inmod))
        df_pt = df_pt.append(pd.read_csv(join("..", dir_id, "pt.csv")).assign(setting="id").assign(inmod=inmod))
        return df_run, df_pt
    else:
        return df_run
```

```python
def save_tex(df, setting, inmod, sex):
    df = df.query("setting == '{}'".format(setting))
    df = df.dropna(0, subset=["ba_mean"])
    df = df[["technique", "io", "model", "ba_mean", "ba_std", "ba_ci_low", "ba_ci_high", "ba_pvalue", "sens", "spec", "auc_mean", "auc_pvalue"]]
    df.model = df.model.map({
        "LogisticRegression" : "LR", 
        "LinearSVC" : "LSVM", 
        "SVC" : "KSVM", 
        "GradientBoostingClassifier" : "GB", 
        "GaussianNB" : "NB"
    })
    df.technique = df.technique.map({
        "baseline" : "baseline", 
        "cb" : "cb", 
        "wdcr" : "wdcr",
        "cvcr": "cvcr",
        "cvcrcb" : "cbcvcr", 
        "cbcvcr" : "cbcvcr"
    })
    df['technique'] = pd.Categorical(df['technique'], categories=["baseline", "cb", "wdcr", "cvcr", "cbcvcr"])
    df['io'] = pd.Categorical(df['io'], categories=["X-y", "X-s", "X-c", "s-y", "c-y"])
    df = df.sort_values(by=["technique", "io"])
    df = df.reset_index(drop=True)

    df.to_latex("{}-{}-{}.tex".format(inmod, setting, sex))
    return df
```
